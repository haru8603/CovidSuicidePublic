# Rotemberg Diagnosis

## Setup

```python
%%stata
qui{
capture restore    
use ../data/suicide_analysis_data_static,clear
xtset citycode covid  

/* Local macro */
local clsvars="citycode"

local gmm2="gmm2s"
local log_="log_"
sum suicide
local suicide_t_m=r(mean)
sum suicide_m
local suicide_m_m=r(mean)
sum suicide_f
local suicide_f_m=r(mean)
sum `log_'active_opening_ratio
local log_active_opening_ratio_m=r(mean)
sum `log_'unemployment_rate
local log_unemployment_rate_m=r(mean)
sum `log_'patient
local log_patient_m=r(mean)
sum log_active_opening_weighted
local log_active_opening_weighted_m=r(mean)
sum log_bankruptcy_debt
local log_bankruptcy_debt_m=r(mean)

* take first difference    
gen D_mobility_index=D.mobility_index
* gen D_stay_home=D.stay_home      
foreach x in u20 20_40 40_60 o60{
  gen D_suicide_f_`x'_log=D.suicide_f_`x'_log  
}
    
    
* shift share make
rename commuter_ratio_time00 share_0
rename commuter_ratio_time30u share_1
rename commuter_ratio_time30_60 share_2
rename commuter_ratio_time60_90 share_3
rename commuter_ratio_time90_120 share_4
rename commuter_ratio_time120o share_5


gen shift_0=telework_ratio_time00_ac-telework_ratio_time00_bc    
gen shift_1=telework_ratio_time30u_ac-telework_ratio_time30u_bc    
gen shift_2=telework_ratio_time30_60_ac-telework_ratio_time30_60_bc
gen shift_3=telework_ratio_time60_90_ac-telework_ratio_time60_90_bc
gen shift_4=telework_ratio_time90_120_ac-telework_ratio_time90_120_bc
gen shift_5=telework_ratio_time120o_ac-telework_ratio_time120o_bc 

gen D_telework_ratio=0    
local variables_shift_share="" 
local variables_iv=""    
foreach x of numlist 0/5{
  gen iv_`x'=share_`x'*shift_`x'    
  replace D_telework_ratio=D_telework_ratio+iv_`x' 
  local variables_shift_share="`variables_shift_share' share_`x' shift_`x'  "     
  local variables_iv="`variables_iv' iv_`x'"    
}   
local variables_orders_all=""     
local variables_orders_all="`variables_shift_share' `variables_iv' D_telework_ratio"
    
}
```

```python
%%stata
qui{
local sfx="u20"
sum suicide_f_`sfx'
local suicide_f_`sfx'_m=r(mean)
local paracomp_f_mobility="(stay_home: _b[D_stay_home]*(1+`suicide_f_`sfx'_m')/(`suicide_f_`sfx'_m')) "
noi: ivreg2 D_suicide_f_`sfx'_log (D_stay_home=D_telework_ratio), gmm2s cluster(`clsvars')
noi: nlcom `paracomp_f_mobility', post
}    
```

```python
%%stata
qui{
keep if covid==1       
local y D_suicide_f_`sfx'_log
local x D_stay_home
local z D_telework_ratio
order `y' `x' `z' `variables_orders_all'    
ivreg2 D_suicide_f_`sfx'_log (D_stay_home=D_telework_ratio), gmm2s cluster(`clsvars')
noi: di " beta"
noi: di _b[D_stay_home]*(1+`suicide_f_`sfx'_m')/(`suicide_f_`sfx'_m')
}
```

```python
%%stata
qui{
local share_stub share_
local shift_stub shift_
local iv_stub iv_

qui reghdfe `x' `z'  , cluster(`clsvars')
scalar Fa = e(F)
noi: di "First Stage F value"
noi: di Fa
}
```

```python
%%stata
qui{
foreach var of varlist `share_stub'* {
    if regexm("`var'", "`share_stub'(.*)") {
    local k = regexs(1) 
    }
    tempvar temp
    qui gen `temp' = `var' * `shift_stub'`k'
    qui reghdfe `x' `temp' , cluster(`clsvars')
    local pi_`k' = _b[`temp']
    qui test `temp'
    scalar Fv = e(F)
    local F_`k' = Fv

    qui reghdfe `y' `temp' ,  cluster(`clsvars')
    local gamma_`k' = _b[`temp']
    drop `temp'
}
}       
```

```python
%%stata
qui{
tempfile temp
preserve
foreach var of varlist `variables_iv' {
  egen var_`var' = sd(`var')
  replace var_`var' = var_`var'^2
}
keep if _n==1
keep var_*
gen id=_n
reshape long  var_iv_,i(id) j(cat)
drop id
save `temp',replace
restore
}
```

```python
%%stata
qui{
drop if share_0==.
noi: bartik_weight, z(`share_stub'*)  weightstub(`shift_stub'*) x(`x') y(`y') 

mat beta = r(beta)*(1+`suicide_f_`sfx'_m')/(`suicide_f_`sfx'_m')
mat alpha = r(alpha)
mat gamma = r(gam)
mat pi = r(pi)
mat G = r(G)
qui desc `share_stub'*, varlist
local varlist = r(varlist)

svmat beta
svmat alpha
svmat gamma
svmat pi
svmat G  
save rotemberg_data,replace
}
```

```python
%%stata
qui{
gen cat = ""
local t = 1
foreach var in `varlist' {
    if regexm("`var'", "`share_stub'(.*)") {
    qui replace cat = regexs(1) if _n == `t'
    }
    local t = `t' + 1
}
destring(cat),replace force
keep  cat beta1* alpha* gamma* pi* G*
drop if cat==.
gen commute_time="zero min"
replace commute_time="      less than 30 min" if cat==1
replace commute_time=" between 30 and 60 min" if cat==2
replace commute_time=" between 60 and 90 min" if cat==3
replace commute_time="between 90 and 120 min" if cat==4
replace commute_time="     more than 120 min" if cat==5
noi: list cat commute_time beta1 alpha1 G1,table clean  noobs  
}
```

```python
%%stata
qui{
gen alpha_sign=-1 if alpha1<0
replace  alpha_sign=0 if alpha1==0
replace  alpha_sign=1 if alpha1>0
label define sign -1 "negative (-)" 0 "0" 1 "positive (+)"
label value alpha_sign sign
noi tab  alpha_sign

gen F = .
gen pi = .
gen gamma = .
levelsof cat, local(cats)
foreach cat in `cats' {
	capture replace F = `F_`cat'' if cat == `cat'
	capture replace pi = `pi_`cat'' if cat == `cat'
	capture replace gamma = `gamma_`cat'' if cat == `cat'
	}
merge 1:1 cat using `temp'
noi: corr alpha1 G1 beta1 F var_iv
save rotemberg_result,replace
gen omega = alpha1*beta1
noi: total omega
mat b = e(b)
global omega = b[1,1]
noi: di $omega      
local b=$omega   
}
```

### Table 2: Panle A

```python
%%stata
qui{
use rotemberg_result,clear
 foreach cat in `cats' {
   sum alpha1 if cat == `cat'
   local alpha_`cat' = string(r(mean),"%9.4f")
   sum G1 if cat == `cat' 
   local G_`cat' = string(r(mean), "%9.4f")
   sum beta1 if cat == `cat'
   local beta_`cat' = string(r(mean),"%9.4f")  
   egen first_commute_time = first(commute_time)  if cat == `cat'
   gsort - first_commute_time 
   list first_commute_time 
   local commute_time_`cat'=first_commute_time[1]
   drop first_commute_time    
}
use rotemberg_data,clear
foreach cat in `cats' {
   tempvar temp
   qui gen `temp' =  `share_stub'`cat' * `shift_stub'`cat'
   ch_weak, p(.05) beta_range(-10.0(.01)5.0)  y(`y') x(`x') z(`temp') cluster(`clsvars')
   local omega_min= r(beta_min)*(1+`suicide_f_`sfx'_m')/(`suicide_f_`sfx'_m')
   local omega_max= r(beta_max)*(1+`suicide_f_`sfx'_m')/(`suicide_f_`sfx'_m')

   local ci_min_`cat' =string(`omega_min', "%20.3f")
   local ci_max_`cat' =string(`omega_max', "%20.3f")
   if `cat'>=1{ 
   if `cat'==1{
   noi: disp  as text "     commuting time  : alpha   g   beta_k   95 % CI  "
   }
   noi: disp "`commute_time_`cat'':  `alpha_`cat''  `G_`cat''  `beta_`cat''   [`ci_min_`cat'', `ci_max_`cat'']"
   }
   drop `temp'
}
}
```

```python
%%stata
qui{

local absvars="citycode covid"
local clsvars="citycode"
local absvars_ind "i.citycode i.covid"    
local gmm2="gmm2s"
local log_="log_"
    
matrix define CF = (0, 0)

* min specification

use ../data/data2,clear
xtset citycode covid

ivreghdfe suicide_f_u20_log (stay_home=commuting_time_iv), gmm2s absorb(`absvars', savefe) resid cluster(`clsvars')  
predict pred_f_iv_min, xbd
gen pred_num_f_iv_min = exp(pred_f_iv_min) - 1    
bysort citycode: gen cf_f_iv_min =  exp(_b[stay_home] * stay_home + __hdfe1__[_n+1] + __hdfe2__[_n+1] ) - 1   
bysort citycode: gen pred_cf_diff_f_iv_min = cf_f_iv_min - pred_num_f_iv_min[_n+1]                                                                                                                                  
sum pred_cf_diff_f_iv_min 
scalar total_pred_cf_diff_f_iv_min = r(sum)
sum pred_num_f_iv_min if covid == 1   
scalar total_pred_num_f_iv_min = r(sum)
sum cf_f_iv_min
scalar total_pred_cf_f_iv_min = r(sum)
noi di total_pred_cf_f_iv_min
noi di total_pred_num_f_iv_min
scalar percent_change_cf_min = (total_pred_cf_f_iv_min - total_pred_num_f_iv_min)/total_pred_cf_f_iv_min
matrix CF[1,1]=percent_change_cf_min*100         
     
* max specification
    
use ../data/data0,clear
xtset citycode covid   
    
ivreghdfe suicide_f_u20_ihs (stay_home=commuting_time60_90_iv) log_active_opening_ratio log_unemployment_rate, gmm2s absorb(`absvars', savefe) resid cluster(`clsvars')  
predict pred_f_iv_max, xbd
gen pred_num_f_iv_max = sinh(pred_f_iv_max)          
bysort citycode: gen cf_f_iv_max =  sinh(_b[stay_home] * stay_home + _b[log_active_opening_ratio] * log_active_opening_ratio + _b[log_unemployment_rate] * log_unemployment_rate + __hdfe1__[_n+1] + __hdfe2__[_n+1] )
bysort citycode: gen pred_cf_diff_f_iv_max = cf_f_iv_max - pred_num_f_iv_max[_n+1]                                                                                                                                  
sum pred_cf_diff_f_iv_max 
scalar total_pred_cf_diff_f_iv_max = r(sum)
sum pred_num_f_iv_max if covid == 1   
scalar total_pred_num_f_iv_max = r(sum)
sum cf_f_iv_max
scalar total_pred_cf_f_iv_max = r(sum)
noi di total_pred_cf_f_iv_max
noi di total_pred_num_f_iv_max
scalar percent_change_cf_max = (total_pred_cf_f_iv_max - total_pred_num_f_iv_max)/total_pred_cf_f_iv_max
matrix CF[1,2]=percent_change_cf_max*100                                                                                                                                         

matrix rownames CF = "Stay-at-home index"
matrix colnames CF = "Min" "Max"
matrix list CF

}

# delimit ;
esttab matrix(CF, fmt(%4.1f)), nomtitles;
```
