# Relationship between the commuting ratio and mobility index

## Female

```python
%%stata
qui{
use ../data/suicide_analysis_data,clear
keep if year==2019|year==2020
gen covid=0
replace covid=1 if (year==2020&month>=4)|year==2021
drop if (year==2019 & month<=3)|(year==2021 & month>=4)
/* mobility adjustment */
replace mobility=0 if mobility<0 & mobility~=.          
collapse (mean) mobility commuter_ratio_f (first) cityname prefecture,by(citycode covid)
#delimit ;    
twoway (scatter mobility commuter_ratio_f if covid==0, msize(tiny)  mc(midblue)) || 
    (scatter mobility commuter_ratio_f if covid==., msize(large)  mc(midblue)) || 
    (lowess mobility commuter_ratio_f if covid==0, lwidth(thick)  lcolor(midblue)) || 
    (scatter mobility commuter_ratio_f if covid==1 , msize(tiny)  mc(navy)) || 
    (scatter mobility commuter_ratio_f if covid==. , msize(large)  mc(navy)) ||     
    (lowess mobility commuter_ratio_f if covid==1, lwidth(thick)  lcolor(navy) lpattern(solid)), 
    legend(label(2 "Before covid")label(5 "After covid") order (2 5)) ytitle(Mobility index) 
    xtitle(Female Commuter Ratio) name(scatter1,replace) graphregion(color(white));
#delimit cr
}

graph save  ../figuretable/figure/female_commuter_mobility_scatter.gph,replace
graph export ../figuretable/figure/female_commuter_mobility_scatter.jpg, width(3200) replace
```

# Suicide rate for the first and forth quantiles of commuting ratios

## Female

```python
%%stata
qui{
use ../data/suicide_analysis_data,clear     
/* mobility adjustment */
replace mobility=0 if mobility<0 & mobility~=.          
collapse (mean) commuter_ratio (sum) suicide suicide_m suicide_f (first)  pop_m pop_f, by(modate commuter_type)
keep if modate<=monthly("2021m12","YM")  & modate>=monthly("2019m1","YM")      
local ltp=" plot1(lc(navy) lpattern(solid)  lwidth(thick)) plot2(lc(navy) lpattern(solid)  lwidth(thick))"           
xtset commuter_type modate
keep if commuter_type==1|commuter_type==4    
gen log_suicide_f=log(suicide_f)
gen suicide_rate_f=suicide_f/pop_f*100000
sum suicide_rate_f
gen max=r(max)*1.2
xtline suicide_rate_f, addplot(area max modate if modate>=monthly("2020m4","YM")& modate<=monthly("2020m5","YM") , color(blue%10) || ///
                              area max modate if modate>=monthly("2021m1","YM")& modate<=monthly("2021m3","YM") , color(blue%10) ) ///
    i(commuter_type) t(modate) overlay name(suicide_rate_f, replace) saving(suicide_rate_f,replace)   ///
    xtitle("Month") xlabel(708 "2019/01" 720 "2020/01" 732 "2021/01" 744 "2022/01") ytitle("Incidence per 100,000") `ltp' ///
    legend(off) text(280 719.5 "{bf:Commuting ratio: 4th quantile}") text(100 715.5 "{bf:Commuting ratio: 1st quantile}") /// 
    graphregion(color(white) )  
}

graph save  ../figuretable/figure/female_suicide_commuer_percentile.gph,replace
graph export ../figuretable/figure/female_suicide_commuer_percentile.jpg, width(3200) replace
```

#  Mobility index

## Trend of MI1

```python
%%stata
qui{
capture restore
use ../data/suicide_analysis_data,clear
drop if year<=2018
local s1=monthly("2019m1","YM")
local s2=monthly("2021m12","YM")
drop if modate>`s2'
/* mobility adjustment */
replace mobility=0 if mobility<0 & mobility~=.        
preserve
collapse (mean) mobility,by(modate)
gen dum=0.4
twoway (line mobility modate, ysc(range(0.2 0.4) ) ylabel(0.2(0.1)0.4) lwidth(thick)) || ///
    (area dum modate if modate>=monthly("2020m4","YM")& modate<=monthly("2020m5","YM") , color(blue%10)) || ///
    (area dum modate if modate>=monthly("2021m1","YM")& modate<=monthly("2021m3","YM") , color(blue%10) ), ///
    legend(off) xtitle("Month") ytitle("Mobility index (Outflow ratio)") xsc(r(`s1' `s2')) name(mobility_trend,replace) ///
    xlabel(708 "2019/01" 720 "2020/01" 732 "2021/01" 744 "2022/01") graphregion(color(white))
}  

graph save ../figuretable/figure/mobility_trend.gph,replace
graph export ../figuretable/figure/mobility_trend.jpg, width(3200) replace
```

## Kernel density of MI1

```python
%%stata
qui{
use ../data/suicide_analysis_data,clear
keep if year==2019|year==2020
keep if year==2019|year==2020
gen covid=0
replace covid=1 if (year==2020&month>=4)|year==2021
drop if (year==2019 & month<=3)|(year==2021 & month>=4)
/* mobility adjustment */
replace mobility=0 if mobility<0 & mobility~=.         
collapse (mean) mobility (first) cityname prefecture,by(citycode covid)
twoway (kdensity mobility if covid==0,lc(midblue) lwidth(thick))|| ///
    (kdensity mobility if covid==1,lc(navy) lwidth(thick)), ///
    legend(label(1 "Before ") label(2 "After ") position(1) ring(0) col(1) region(lstyle(none))) ///
    ytitle("Density") xtitle("Mobility index (Outflow ratio)") name(mobility_distribution,replace) graphregion(color(white))

}
graph save ../figuretable/figure/mobility_distribution.gph,replace
graph export ../figuretable/figure/mobility_distribution.jpg, width(3200) replace
```

# Mobility data, examples of three cities

```python
%%stata

qui{
capture restore   
tempfile temp
use ../data/from_to_population_flow_hour_agoop,clear
drop timezone
destring month,force replace
reshape wide population,i(year month prefcode citycode hour) j(from_area)
destring month,force replace
preserve
    
* 港区
keep if citycode=="13103"&year==2019&month==4  
gen population12=population1+population2
gen log_population0=log(population0)    
gen log_population12=log(population12)  
save `temp',replace
gen dum=0 if hour>=1 & hour<=4
replace dum=1 if hour>=11 & hour<=14
drop if dum==.    
collapse (sum) population0,by(dum)
local  mobility= 1 - round(population0[2]/population0[1],0.01)
restore
preserve
keep if citycode=="13103"&year==2019&month==4  
sum population0
local cv = round(r(sd)/r(mean), 0.01)
use `temp',clear    
graph bar population0 population12, over(hour, label(alternate labsize(vsmall))) stack name(bar1,replace) ///
    legend(order(2 1) label(1 "Residents") label(2 "Non-residents")  position(6) col(1) region(lstyle(none)) nobox) ///
    bar(1, fcolor("106 208 200") lwidth(none)) bar(2, fcolor(orange) lwidth(none)) ///
    title("Minato, Tokyo"  "(Business district)"  "Outflow ratio=`mobility'" "CV=`cv'") graphregion(color(white)) nodraw
restore
preserve
    
* 東京都国分寺市
keep if citycode=="13214"&year==2019&month==4  
gen population12=population1+population2
gen log_population0=log(population0)    
gen log_population12=log(population12)    
save `temp',replace
gen dum=0 if hour>=1 & hour<=4
replace dum=1 if hour>=11 & hour<=14
drop if dum==.    
collapse (sum) population0,by(dum)
local  mobility= 1 - round(population0[2]/population0[1],0.01)
restore
preserve
keep if citycode=="13214"&year==2019&month==4  
sum population0
local cv = round(r(sd)/r(mean), 0.01)
use `temp',clear        
graph bar population0 population12,over(hour,label(alternate labsize(vsmall))) stack  name(bar2,replace) ///
    legend(order(2 1) label(1 "Residents") label(2 "Non-residents")  position(6) col(1)  region(lstyle(none)) nobox) ///
    bar(1, fcolor("106 208 200") lwidth(none)) bar(2, fcolor(orange) lwidth(none)) ///
    title("Kokubunji, Tokyo"  "(Residential area)"  "Outflow ratio=`mobility'" "CV=`cv'") graphregion(color(white)) nodraw ///
    ysc(r(0,130000)) ylab(0(25000)130000)
restore
preserve
    
* 沖縄県石垣市
keep if citycode=="47207"&year==2019&month==4  
gen population12=population1+population2
gen log_population0=log(population0)    
gen log_population12=log(population12) 
save `temp',replace
gen dum=0 if hour>=1 & hour<=4
replace dum=1 if hour>=11 & hour<=14
drop if dum==.    
collapse (sum) population0,by(dum)
local  mobility= 1 - round(population0[2]/population0[1],0.01)
restore
preserve
keep if citycode=="47207"&year==2019&month==4  
sum population0
local cv = round(r(sd)/r(mean), 0.01)
use `temp',clear        
graph bar population0 population12,over(hour,label(alternate labsize(vsmall))) stack  name(bar3,replace) ///
    legend(off) ///
    bar(1, fcolor("106 208 200") lwidth(none)) bar(2, fcolor(orange) lwidth(none)) ///
    title("Ishigaki, Okinawa" "(Island area)"  "Outflow ratio=`mobility'" "CV=`cv'")  graphregion(color(white)) nodraw
restore
grc1leg bar1 bar2 bar3, legendfrom(bar2) name(foot_trafic,replace) saving(bars,replace) col(3) graphregion(color(white)) 
}

graph save ../figuretable/appendix/foot_trafic_mobility_index.gph,replace
graph export ../figuretable/appendix/foot_trafic_mobility_index.jpg, width(3200) replace
```

# Within-city mobility index

## Kernel Density

```python
%%stata
local mb=0

qui{
use ../data/suicide_analysis_data,clear
merge 1:1 citycode year month using ../data/station_place_cv_mb`mb'
drop if _merge==2    
drop _merge       
merge 1:1 citycode year month using ../data/place_cv_mb`mb'    
drop if _merge==2
drop _merge  
merge 1:1 citycode year month using ../data/station_cv_mb`mb'    
drop if _merge==2
drop _merge      
drop if year<=2018
keep if year==2019|year==2020
gen covid=0
replace covid=1 if (year==2020&month>=4)|year==2021
drop if (year==2019 & month<=3)|(year==2021 & month>=4)
collapse (mean) cv_residents_place (first) cityname prefecture, by(citycode covid)
twoway (kdensity cv_residents_place if covid==0,lc(midblue) lwidth(thick))|| ///
    (kdensity cv_residents_place if covid==1,lc(navy) lwidth(thick)), ///
    legend(label(1 "Before ") label(2 "After ") position(1) ring(0) col(1)  region(lstyle(none))) ///
    ytitle("Density") xtitle("Mobility index (Within-city CV)") name(cv_residents_distribution,replace) graphregion(color(white))
}

graph save ../figuretable/appendix/cv_residents_within_city_distribution.gph,replace
graph export ../figuretable/appendix/cv_residents_within_city_distribution.jpg, width(3200) replace
```

# Correlation

```python
%%stata
local mb=0
qui{
use ../data/suicide_analysis_data,clear
merge 1:1 citycode year month using ../data/station_place_cv_mb`mb'
drop if _merge==2    
drop _merge       
merge 1:1 citycode year month using ../data/place_cv_mb`mb'    
drop if _merge==2
drop _merge  
merge 1:1 citycode year month using ../data/station_cv_mb`mb'    
drop if _merge==2
drop _merge      
drop if year<=2018
keep if year==2019|year==2020
gen covid=0
replace covid=1 if (year==2020&month>=4)|year==2021
drop if (year==2019 & month<=3)|(year==2021 & month>=4)
    
collapse (sum) suicide*  population  pop_m pop_f state_emergency_days state_emergency_1_days state_emergency_2_days school_closure_days  (mean) mobility stay_home *commuter_ratio*  active_opening_ratio active_opening_ratio_weighted unemployment_rate prefcode  cv_* (first) prefecture cityname,by(citycode covid)

xtset citycode covid    

twoway scatter  D.cv_residents_place  D.mobility if D.cv_residents_place~=. &   D.mobility~=. , msymbol(o) msize(tiny) || /// 
lfit D.cv_residents_place  D.mobility  if D.cv_residents_place~=. &   D.mobility~=. , lcolor(navy) lwidth(0.8) ///
legend(off) ytitle("ΔMobility index (Within-city CV)") xtitle("ΔMobility index (Outflow ratio)") name(sc_place,replace)  graphregion(color(white))
}

graph save ../figuretable/appendix/cv_residents_correlation.gph,replace
graph export ../figuretable/appendix/cv_residents_correlation.jpg, width(3200) replace
```
